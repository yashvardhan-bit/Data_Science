```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=sns.load_dataset("tips")
data
```

```python
plt.plot(data['total_bill'])
plt.show
```

```python
sns.lineplot(data['total_bill'])
```

```python
sns.scatterplot(data['total_bill'])
```

```python
sns.scatterplot(x='total_bill', y='tip', data=data)
```

```python
sns.histplot(data['total_bill'])
```

```python
sns.kdeplot(data['total_bill'])
```

```python
#catplot
#relplot
```

```python
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', data=data, hue='time')
```

```python
sns.relplot(x='total_bill', y='tip', data=data, hue='time',style='smoker')
```

```python
sns.relplot(x="total_bill",y="tip",data=data,hue="day",kind="scatter",row="time",style="smoker")
```

```python
# This creates a scatter plot of 'total_bill' vs 'tip' using seaborn's relplot,
# with separate rows for each 'time' (Lunch/Dinner) and columns for each 'smoker' status.
sns.relplot(x='total_bill', y='tip', data=data, row="time", col='smoker',  kind='scatter')
```

```python
sns.relplot(x='total_bill', y='tip', data=data, row="day", col='smoker',  kind='scatter')
```

```python
sns.relplot(x='total_bill', y='tip', data=data, hue="time", col='day',  kind='scatter')
```
