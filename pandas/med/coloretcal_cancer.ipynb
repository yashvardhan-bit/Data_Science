```python
df.drop_duplicates(inplace=True)
df
```

```python
gen_dict = {'M':1, 'F':2}
df["Gender"]= df['Gender'].map(gen_dict)
```

```python
df['Cancer_Stage'].unique()
```

```python
cs_dict = {'Localized':1, 'Regional':2, 'Metastatic':3}
df["Cancer_Stage"]= df['Cancer_Stage'].map(cs_dict)
```

```python
df['Family_History'].unique()
```

```python
fh_dict = {'No':0, 'Yes':1}
df["Family_History"]= df['Family_History'].map(fh_dict)
```

```python
df['Alcohol_Consumption'].unique()
```

```python
ac_dict = {'No':0, 'Yes':1}
df["Alcohol_Consumption"]= df['Alcohol_Consumption'].map(ac_dict)
```

```python
df
```

```python
df['Obesity_BMI'].unique()
```

```python
bmi_dict = {'Overweight':1, 'Normal':2, 'Obese':3}
df["Obesity_BMI"]= df['Obesity_BMI'].map(bmi_dict)
```

```python
df['Survival_5_years'].unique()
```

```python
sur_dict = {'Yes':1, 'No':0}
df["Survival_5_years"]= df['Survival_5_years'].map(sur_dict)
```

```python
df['Mortality'].unique()
```

```python
mor_dict = {'Yes':1, 'No':0}
df["Mortality"]= df['Mortality'].map(mor_dict)
```

```python
df['Survival_Prediction'].unique()
```

```python
sur_p_dict = {'Yes':1, 'No':0}
df["Survival_Prediction"]= df['Survival_Prediction'].map(sur_p_dict)
```

```python
df['Healthcare_Access'].unique()
```

```python
hlt_dict = {'Moderate':1, 'High':2, 'Low':0}
df["Healthcare_Access"]= df['Healthcare_Access'].map(hlt_dict)
```

```python
df['Survival_Prediction'].unique()
```

```python
sur_p_dict = {'Yes':1, 'No':0}
df["Survival_Prediction"]= df['Survival_Prediction'].map(sur_p_dict)
```

```python
df
```

```python
df.drop(['Country', 'Urban_or_Rural', 'Economic_Classification', 'Insurance_Status'],axis=1, inplace=True)
df
```
